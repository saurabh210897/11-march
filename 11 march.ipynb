```python
# Q1: What is the difference between a t-test and a z-test? Provide an example scenario where you would 
# use each type of test.
```

```python
# Both t-test and z-test are statistical tests used to test hypotheses about population means based on sample data. 
# However, they differ in their assumptions about the population variance and sample size.

# A t-test is used when the population variance is unknown and must be estimated from the sample data. It is also preferred when the sample size is small (typically less than 30). 
# A t-test uses a t-distribution to calculate the test statistic and determine the p-value.

# Example scenario: Suppose you want to compare the mean height of two groups of students in a high school, one group consisting of boys and the other consisting of girls. 
# You have collected a sample of heights from each group, and you want to test whether there is a significant difference in the mean height between the two groups. 
# In this case, you would use a two-sample t-test.

# A z-test is used when the population variance is known or when the sample size is large (typically greater than 30). 
# A z-test uses a standard normal distribution to calculate the test statistic and determine the p-value.

# Example scenario: Suppose you are testing a new weight loss supplement and want to compare the mean weight loss of a group of participants who took the supplement 
# with the population mean weight loss of people who do not take the supplement. You have a large sample size and know the population standard deviation. 
# In this case, you would use a z-test.
```

```python
# Q2: Differentiate between one-tailed and two-tailed tests.
```

```python
# A one-tailed test, also known as a directional test, is a hypothesis test in which the null hypothesis is rejected if the sample mean is significantly greater than or less than 
# the hypothesized population mean in a specific direction. The alternative hypothesis specifies the direction of the difference. 
# The rejection region is located entirely in one tail of the distribution. One-tailed tests are used when the researcher has a specific hypothesis about
# the direction of the difference between the sample mean and population mean.

# For example, suppose you want to test whether a new medication reduces blood pressure. The null hypothesis would be that the medication has no effect on blood pressure, 
# and the alternative hypothesis would be that the medication reduces blood pressure. A one-tailed test would be used to determine whether the medication reduces blood pressure 
# in a specific direction, i.e., the alternative hypothesis specifies that the medication reduces blood pressure.

# A two-tailed test, also known as a non-directional test, is a hypothesis test in which the null hypothesis is rejected if the sample mean is significantly different from 
# the hypothesized population mean in either direction. The alternative hypothesis does not specify the direction of the difference. The rejection region is located in 
# both tails of the distribution. Two-tailed tests are used when the researcher is interested in determining whether there is a difference between the sample mean and population mean,
# regardless of the direction.

# For example, suppose you want to test whether a coin is fair. The null hypothesis would be that the coin is fair, and the alternative hypothesis would be that the coin is not fair.
# A two-tailed test would be used to determine whether the coin is biased in either direction, i.e., the alternative hypothesis does not specify a direction.
```

```python
# Q3: Explain the concept of Type 1 and Type 2 errors in hypothesis testing. Provide an example scenario for 
# each type of error.
```

```python
# Type 1 and Type 2 errors are two types of errors that can occur in hypothesis testing.

# Type 1 error, also known as a false positive, occurs when the null hypothesis is rejected when it is actually true. In other words,
# the researcher concludes that there is a significant difference between the sample mean and population mean when, in fact, there is no difference. 
# Type 1 errors are considered to be more serious than Type 2 errors because they can lead to incorrect conclusions and wasted resources.

# Example scenario: Suppose a company wants to test a new drug's effectiveness in reducing pain. The null hypothesis is that the drug has no effect on pain reduction. 
# If the company rejects the null hypothesis and concludes that the drug is effective, but, in reality, the drug has no effect, it would be a Type 1 error.

# Type 2 error, also known as a false negative, occurs when the null hypothesis is not rejected when it is actually false. 
# In other words, the researcher concludes that there is no significant difference between the sample mean and population mean when, in fact, there is a difference. 
# Type 2 errors are less serious than Type 1 errors because they do not lead to incorrect conclusions, but they may result in missed opportunities for further investigation or improvement.

# Example scenario: Suppose a company wants to test a new drug's effectiveness in reducing pain. The null hypothesis is that the drug has no effect on pain reduction. 
# If the company fails to reject the null hypothesis and concludes that the drug is not effective, but, in reality, the drug is effective, it would be a Type 2 error.

# In hypothesis testing, the researcher must balance the risk of making Type 1 and Type 2 errors. 
# The level of significance (alpha) and power of the test are two factors that influence the probability of making these errors. 
# By setting a lower alpha level, the risk of Type 1 error can be reduced, but it increases the risk of Type 2 error. 
# By increasing the power of the test, the risk of Type 2 error can be reduced, but it increases the risk of Type 1 error.
```

```python
# Q4:  Explain Bayes's theorem with an example.
```

```python
# Bayes's theorem is a mathematical formula used to update probabilities based on new evidence. It provides a way to calculate the probability of an event occurring, 
# given prior knowledge and new information. The formula is as follows:

# P(A|B) = P(B|A) x P(A) / P(B)

# where P(A|B) is the conditional probability of event A given event B, P(B|A) is the conditional probability of event B given event A, 
# P(A) is the prior probability of event A, and P(B) is the prior probability of event B.

# An example scenario to illustrate the application of Bayes's theorem is as follows:

# Suppose a medical test for a rare disease is known to be 98% accurate. That is, the test has a 98% sensitivity (ability to correctly identify true positive cases) and 
# a 98% specificity (ability to correctly identify true negative cases). The prevalence of the disease in the population is 1 in 1000.

# Now, imagine a person takes the test and tests positive. The question is, what is the probability that the person actually has the disease?

# Using Bayes's theorem, we can calculate the probability as follows:

# Let A be the event of having the disease, and B be the event of testing positive for the disease.

# P(A) = 1/1000 (prior probability of having the disease)
# P(B|A) = 0.98 (sensitivity, probability of testing positive given that the person has the disease)
# P(B|not A) = 0.02 (1 - specificity, probability of testing positive given that the person does not have the disease)

# To calculate P(B), we need to use the law of total probability, which states that the probability of an event occurring is equal to 
# the sum of the probabilities of that event occurring under each possible condition. In this case:

# P(B) = P(A) x P(B|A) + P(not A) x P(B|not A)
# = (1/1000) x 0.98 + (999/1000) x 0.02
# = 0.02176

# Using Bayes's theorem:

# P(A|B) = P(B|A) x P(A) / P(B)
# = 0.98 x (1/1000) / 0.02176
# = 0.045

# Therefore, the probability that the person actually has the disease given a positive test result is only 4.5%. 
# This example shows that even with a highly accurate test, the probability of a positive test result indicating the presence of the disease 
# is low when the disease is rare in the population.
```

```python
# Q5: What is a confidence interval? How to calculate the confidence interval, explain with an example.
```

```python
# A confidence interval is a range of values that is likely to contain the true population parameter with a certain level of confidence.
# It is a statistical measure that quantifies the uncertainty associated with a sample estimate.

# The confidence interval is calculated using the sample mean, standard deviation, sample size, and the level of confidence desired. 
# The level of confidence is usually expressed as a percentage and represents the probability that the true population parameter falls within the interval.
# For example, a 95% confidence interval means that if the sampling process were repeated many times, 95% of the intervals obtained would contain the true population parameter.

# The formula for calculating the confidence interval is:

# Confidence interval = sample mean ± (critical value x standard error)

# where the critical value is obtained from a statistical table based on the level of confidence desired and the degrees of freedom (n-1), 
# and the standard error is calculated as the standard deviation of the sample divided by the square root of the sample size.

# An example scenario to illustrate the calculation of a confidence interval is as follows:

# Suppose a company wants to estimate the mean age of its employees. A random sample of 50 employees is selected, and their ages are recorded. 
# The sample mean is 35 years, and the sample standard deviation is 5 years. The company wants to calculate a 95% confidence interval for the true population mean age.

# The first step is to calculate the standard error:

# Standard error = standard deviation / √sample size
# = 5 / √50
# = 0.707

# The next step is to determine the critical value from the statistical table. For a 95% confidence interval with 49 degrees of freedom, the critical value is 2.009.

# Finally, the confidence interval can be calculated as:

# Confidence interval = sample mean ± (critical value x standard error)
# = 35 ± (2.009 x 0.707)
# = (33.40, 36.60)

# Therefore, with 95% confidence, we can say that the true population mean age of the company's employees falls within the range of 33.40 to 36.60 years.
```

```python
# Q6. Use Bayes' Theorem to calculate the probability of an event occurring given prior knowledge of the 
# event's probability and new evidence. Provide a sample problem and solution
```

```python
# Suppose a factory produces light bulbs, and 5% of the bulbs are defective. A quality control inspector tests a bulb and finds that it is defective.
# What is the probability that the bulb was produced by a particular machine that is known to be more likely to produce defective bulbs, 
# given that 70% of the defective bulbs come from this machine?

# Let A be the event that the bulb was produced by the particular machine, and B be the event that the bulb is defective. We want to find P(A|B), 
# the probability that the bulb was produced by the machine given that it is defective.

# We know that:

# P(A) = 1/4 (there are four machines producing the bulbs, so the prior probability of the bulb being produced by the machine in question is 1/4)
# P(B|A) = 0.7 (the probability of a defective bulb given that it was produced by the machine in question)
# P(B|not A) = 0.05 (the probability of a defective bulb given that it was not produced by the machine in question)
# P(not A) = 3/4 (the prior probability that the bulb was not produced by the machine in question)
# Using Bayes' theorem, we can calculate the probability we're interested in:

# P(A|B) = P(B|A) * P(A) / P(B)
# = P(B|A) * P(A) / (P(B|A) * P(A) + P(B|not A) * P(not A))
# = 0.7 * (1/4) / (0.7 * (1/4) + 0.05 * (3/4))
# ≈ 0.536

# Therefore, given that the bulb is defective, the probability that it was produced by the particular machine is about 53.6%.
```

```python
# Q8. What is the margin of error in a confidence interval? How does sample size affect the margin of error? 
# Provide an example of a scenario where a larger sample size would result in a smaller margin of error.
```

```python
# The margin of error in a confidence interval is the range of values that we add and subtract to the point estimate of a parameter (such as the population mean) 
# to create an interval that captures the true value of the parameter with a certain degree of confidence. It represents the precision or accuracy of the estimate.

# The margin of error is influenced by several factors, including the level of confidence desired, the sample size, and 
# the standard deviation of the population (or the standard error of the mean). Generally, as the level of confidence increases, the margin of error increases,
# while as the sample size or the standard deviation decreases, the margin of error decreases.

# A larger sample size generally results in a smaller margin of error, as it reduces the variability and uncertainty associated with the estimate. 
# This is because larger sample sizes provide more information about the population and reduce the sampling error (the difference between the sample mean and the true population mean). 
# As a result, the standard error of the mean decreases as the sample size increases, leading to a smaller margin of error.

# For example, suppose we want to estimate the average height of all college students in a particular state with a 95% confidence interval and a margin of error of 1 inch. 
# If we take a sample of 100 students and find a sample mean of 68 inches and a sample standard deviation of 3 inches, we can calculate the margin of error as:

# Margin of error = 1.96 x (3 / √100) ≈ 0.588

# Thus, the 95% confidence interval for the population mean would be:

# Confidence interval = 68 ± 0.588 = (67.41, 68.59)

# Now, if we increase the sample size to 400 students and assume the same sample mean and standard deviation, the margin of error would be:

# Margin of error = 1.96 x (3 / √400) ≈ 0.294

# Thus, the 95% confidence interval for the population mean would be:

# Confidence interval = 68 ± 0.294 = (67.71, 68.29)

# As we can see, the larger sample size results in a smaller margin of error and a narrower confidence interval, indicating a more precise estimate of the population mean.
```

```python
# Q9. Calculate the z-score for a data point with a value of 75, a population mean of 70, and a population 
# standard deviation of 5. Interpret the results.
```

```python
# To calculate the z-score for a data point with a value of 75, a population mean of 70, and a population standard deviation of 5, we use the formula:

# z = (x - μ) / σ

# where x is the data point, μ is the population mean, and σ is the population standard deviation.

# Plugging in the values, we get:

# z = (75 - 70) / 5 = 1

# The z-score of 1 means that the data point of 75 is one standard deviation above the population mean of 70. In other words,
# the data point is higher than the average by an amount that is typical for this population.

# Z-scores are useful for standardizing data, as they allow us to compare values from different populations or distributions on the same scale.
# They also enable us to calculate probabilities and determine how extreme or unusual a particular data point is relative to the rest of the distribution.
```

```python
# Q10. In a study of the effectiveness of a new weight loss drug, a sample of 50 participants lost an average 
# of 6 pounds with a standard deviation of 2.5 pounds. Conduct a hypothesis test to determine if the drug is 
# significantly effective at a 95% confidence level using a t-test.
```

```python
# To conduct a hypothesis test to determine if the new weight loss drug is significantly effective at a 95% confidence level using a t-test, we need to define the null 
# and alternative hypotheses, choose a significance level, calculate the test statistic and the p-value, and make a decision based on the results.

# Null hypothesis: The new weight loss drug is not significantly effective, and the population mean weight loss is equal to or less than 0 pounds (μ <= 0).
# Alternative hypothesis: The new weight loss drug is significantly effective, and the population mean weight loss is greater than 0 pounds (μ > 0).

# Significance level: α = 0.05 (or 95% confidence level)

# We will use a one-sample t-test, as we have a sample of 50 participants and do not know the population standard deviation. The test statistic for a one-sample t-test is given by:

# t = (x̄ - μ) / (s / √n)

# where x̄ is the sample mean, μ is the hypothesized population mean (in this case, 0), s is the sample standard deviation, and n is the sample size.

# Plugging in the values, we get:

# t = (6 - 0) / (2.5 / √50) = 13.42

# The degrees of freedom for the t-distribution are df = n - 1 = 49. Using a t-table or a calculator, we find that the p-value for a one-tailed test 
# with df = 49 and t = 13.42 is much less than 0.001.

# Since the p-value is less than the significance level of 0.05, we reject the null hypothesis and conclude that the new weight loss drug is significantly effective at 
# a 95% confidence level. This means that we have strong evidence to support the claim that the drug leads to weight loss in the population.

# In summary, based on the sample of 50 participants, the t-test showed that the new weight loss drug is significantly effective, and the population mean weight loss
# is greater than 0 pounds at a 95% confidence level.
```

```python
# Q11. In a survey of 500 people, 65% reported being satisfied with their current job. Calculate the 95% 
# confidence interval for the true proportion of people who are satisfied with their job.
```

```python
# To calculate the 95% confidence interval for the true proportion of people who are satisfied with their job, we can use the following formula:

# CI = p ± Zα/2 * sqrt((p*(1-p))/n)

# where:
# p = sample proportion
# n = sample size
# Zα/2 = the z-score corresponding to the desired confidence level, α/2. For a 95% confidence level, α/2 = 0.025 and the corresponding z-score is 1.96.

# Substituting the values we have:

# p = 0.65 (sample proportion)
# n = 500 (sample size)
# Zα/2 = 1.96 (z-score for a 95% confidence level)

# CI = 0.65 ± 1.96 * sqrt((0.65*(1-0.65))/500)
# CI = 0.65 ± 0.049
# CI = (0.601, 0.699)

# Therefore, we can be 95% confident that the true proportion of people who are satisfied with their job is between 60.1% and 69.9%.
```

```python
# Q12. A researcher is testing the effectiveness of two different teaching methods on student performance. 
# Sample A has a mean score of 85 with a standard deviation of 6, while sample B has a mean score of 82 
# with a standard deviation of 5. Conduct a hypothesis test to determine if the two teaching methods have a 
# significant difference in student performance using a t-test with a significance level of 0.01.
```

```python
# To determine if there is a significant difference in student performance between the two teaching methods, 
# we can conduct a two-sample t-test with the following null and alternative hypotheses:

# Null hypothesis (H0): The mean score of sample A is equal to the mean score of sample B.
# Alternative hypothesis (Ha): The mean score of sample A is not equal to the mean score of sample B.
# We can use a significance level of 0.01, which means we are willing to accept a 1% chance of rejecting the null hypothesis when it is actually true.

# The test statistic for a two-sample t-test is calculated as:

# t = (x1 - x2) / sqrt((s1^2/n1) + (s2^2/n2))

# where:
# x1 = mean score of sample A
# x2 = mean score of sample B
# s1 = standard deviation of sample A
# s2 = standard deviation of sample B
# n1 = sample size of sample A
# n2 = sample size of sample B

# Substituting the given values, we get:

# x1 = 85
# x2 = 82
# s1 = 6
# s2 = 5
# n1 = n2 = sample sizes are not given, assuming n1 = n2 = 30 (recommended minimum sample size for t-test)

# t = (85 - 82) / sqrt((6^2/30) + (5^2/30))
# t = 2.31

# The degrees of freedom for the t-test is calculated as:

# df = n1 + n2 - 2
# df = 58

# Using a t-table or a calculator, we can find the critical t-value for a two-tailed test with a significance level of 0.01 and 58 degrees of freedom to be approximately 2.660.

# Since the calculated t-value (2.31) is less than the critical t-value (2.660), we fail to reject the null hypothesis. 
# Therefore, we do not have enough evidence to conclude that there is a significant difference in student performance between the two teaching methods at a significance level of 0.01.
```

```python
# Q13. A population has a mean of 60 and a standard deviation of 8. A sample of 50 observations has a mean 
# of 65. Calculate the 90% confidence interval for the true population mean.
```

```python
# To calculate the 90% confidence interval for the true population mean, we can use the following formula:

# CI = x̄ ± Zα/2 * (σ/√n)

# where:
# x̄ = sample mean
# Zα/2 = the z-score corresponding to the desired confidence level, α/2. For a 90% confidence level, α/2 = 0.05 and the corresponding z-score is 1.645.
# σ = population standard deviation
# n = sample size

# Substituting the given values, we get:

# x̄ = 65 (sample mean)
# Zα/2 = 1.645 (z-score for a 90% confidence level)
# σ = 8 (population standard deviation)
# n = 50 (sample size)

# CI = 65 ± 1.645 * (8/√50)
# CI = 65 ± 1.854
# CI = (63.146, 66.854)

# Therefore, we can be 90% confident that the true population mean is between 63.146 and 66.854.
```
